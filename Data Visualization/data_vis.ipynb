```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.pyplot import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
np.set_printoptions(precision=4)
```

```python
x = np.arange(-10, 11)
```

```python
fig, axes = plt.subplots(figsize=(8, 6))
axes.plot(x, x**2, color="red", linewidth="2", marker="*", markersize="10", label="X^2")
axes.plot(x, -1 * (x**2), "--b", color="black", label='Inverse of x')
axes.legend('X^2',)
```

```python
x = range(1, 10)
y = [1,2,3,4,0,4,3,2,1]
plt.plot(x,y)
```

```python
plt.bar(x, y)
```

```python
plt.boxplot(y, showmeans=True, showbox=True, vert=False, widths=0.3)
```

```python
z = [1, 2, 3, 4, 0, 4, 3, 2, 1]
fig = plt.figure()
axes = fig.add_axes([.1, .1, 1, 1])
axes.plot(z, lw=2, ds='steps')

axes.set_xlabel('This is x-axis')
axes.set_ylabel('Y-axis')

axes.set_title('My first step plot')

axes.set_label('X-axis')
axes.legend(z,loc='best')

x_ticks = ['1st', '2nd', '3rd', '4th', '5th', '6th', '7th', '8th', '9th']
axes.set_xticklabels(x_ticks, rotation=60, fontsize='medium')

r = np.random.rand(36)

axes2 = fig.add_axes([1, .1, 1, 1])
axes2.plot(r)
```

```python
sns.histplot(z)
```
